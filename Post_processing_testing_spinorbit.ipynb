Now we make equivalent but not equal matrices from our wave-functions:
$$    
\left\langle \Psi_i | {\bf R} | \Psi_j \right\rangle = D_{ij}
$$
Where $\Psi_i$ and $\Psi_j$ are band wave-functions the partners of their corresponding irreps, ${\bf R}$ is symmetry operator and $D_{ij}$ is the matrix of the irrep. Spin wave-functions are represented as:

$$
\Psi = \psi_1 | \alpha\right\rangle + \psi_1 | \beta \right\rangle
$$

where $\psi_1$ and $\psi_2$ are coordinate dependent spin-up and spin-down wavefunctions, respectively and $|\alpha\right\rangle = (1\quad 0)^\dagger$ and $\beta \right\rangle = (0\quad 1)^\dagger$ are spin-up and spin-down operators in standard spin-representation.

```python
def Compute_Perturbation_mat_so(G_vec,C_terms_up,C_terms_dn,
                                ext_basis_up,ext_basis_dn,
                                E_terms,E_ext,m,n):
        new_basis_up = np.vstack((ext_basis_up,C_terms_up))
        new_basis_dn = np.vstack((ext_basis_dn,C_terms_dn))
        dim = len(ext_basis_up)
        cgc = 0
        mat = np.zeros((3,3),dtype=complex)
        Eavg = (E_terms[m]+E_terms[n])/2
        for l in range(len(ext_basis_up)):
            delta_E1 =1/(Eavg - E_ext[l])
            kpi_up = Compute_Matrix_Element(G_vec,new_basis_up,dim+m,l)
            kpi_dn = Compute_Matrix_Element(G_vec,new_basis_dn,dim+m,l)
            kpj_up = Compute_Matrix_Element(G_vec,new_basis_up,l,dim+n)
            kpj_dn = Compute_Matrix_Element(G_vec,new_basis_dn,l,dim+n)
            for i in range(3):
                for j in range(3):
                    uu = kpi_up[i]*kpj_up[j]
                    dd = kpi_dn[i]*kpj_dn[j]
                    ud = kpi_up[i]*kpj_dn[j]
                    du = kpi_dn[i]*kpj_up[j]
                    mat_ij = uu+dd+ud+du
                    mat[i,j] += mat_ij*delta_E1
        return mat
```
